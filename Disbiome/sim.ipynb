```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
disease_embed = pd.read_csv('disease_embeddings.csv',index_col=0)
sim_matrix = cosine_similarity(disease_embed)

df_dis_sim = pd.DataFrame(sim_matrix, index=disease_embed.index, columns=disease_embed.index)

df_dis_sim.to_csv('dd_sim.txt', sep='\t', header=False, index=False)
```

```python
micro_embed = pd.read_csv('microbe_embeddings.csv',index_col=0)
sim_matrix = cosine_similarity(micro_embed)

df_micro_sim = pd.DataFrame(sim_matrix, index=micro_embed.index, columns=micro_embed.index)

df_micro_sim.to_csv('mm_sim.txt', sep='\t', header=False, index=False)
```
